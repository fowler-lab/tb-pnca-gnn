```python
history = pd.read_csv('../data/training_history.csv')
```

```python
def plot_curve(metric, xlabel=False):
    plt.figure(figsize=(6, 4.5))
    plt.plot(history['_step'][:1120], history[f'Train {metric}'][:1120], label=f'Train {metric}')
    plt.plot(history['_step'][:1120], history[f'Test {metric}'][:1120], label=f'Test {metric}')

    plt.gca().spines['top'].set_color('white')
    plt.gca().spines['right'].set_color('white')
    plt.gca().spines['left'].set_color('gray')
    plt.gca().spines['bottom'].set_color('gray')

    if xlabel:
        plt.xlabel('Step')
        plt.gcf().set_size_inches(6, 5)
    else:
        plt.xlabel('')
        plt.xticks([])
    plt.ylabel(metric)
    # plt.title('Training vs Test Loss')
    plt.legend(frameon=False)
    plt.tight_layout()

    plt.show()
```

```python
metrics = ['Loss', 'Accuracy', 'F1', 'Sensitivity', 'Specificity']
for i, metric in enumerate(metrics):
    plot_curve(metric, xlabel=(metric in (['Sensitivity', 'Specificity'])))
```
